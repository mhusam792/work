```python
import pandas as pd
import numpy as np
```

```python
columns = [
    "رقم البوستر", "عدد البيفوتات", "الشركة", "رقم البيفوت",
    "101 مساحة البيفوت", "70 مساحة البيفوت",
    "موقف الجسات تم", "موقف الجسات متبقى",
    "الصلاحية A", "الصلاحية B", "الصلاحية C", "الصلاحية D", "الصلاحية محول تنقيط",
    "المساحة المنزرعة للبيفوت", "مساحه غير الصالح", "مساحة المحول تنقيط", "موقف التسويات", "قواعد البيفوت",
    "عدد أجهزة الري المحوري", "تركيب أجهزة الري المحوري", "استلام الإشراف الزراعي"
]
```

```python
zone3 = pd.read_excel(r"C:\Users\HP\Desktop\البندارى\بيفوتات\fix_pugs\المرحله الاولي.xlsx",
                      sheet_name="zone3",
                      names=columns)
zone3.head(20)
```

```python
cols_to_fill = ['رقم البوستر', 'عدد البيفوتات', 'الشركة']
zone3[cols_to_fill] = zone3[cols_to_fill].replace(['', '–', ' '], np.nan)
zone3[cols_to_fill] = zone3[cols_to_fill].ffill()
zone3.head(20)
```

```python
cols_to_fill_right = ["الصلاحية A", "الصلاحية B", "الصلاحية C", "الصلاحية D"]
zone3[cols_to_fill_right] = zone3[cols_to_fill_right].apply(lambda row: row.ffill(), axis=1)
zone3.head(20)
```

```python
zone3['مساحة البيفوت'] = zone3['101 مساحة البيفوت'].apply(lambda x: 101 if x == 1 else 70)
new_data = zone3[['رقم البوستر', 'عدد البيفوتات', 'الشركة', 'رقم البيفوت',
    'مساحة البيفوت', 'الصلاحية محول تنقيط', 'المساحة المنزرعة للبيفوت',
    'مساحه غير الصالح', 'مساحة المحول تنقيط', 'موقف التسويات','قواعد البيفوت',
    'تركيب أجهزة الري المحوري']]
new_data.head()
```

```python
new_data.to_excel("zone3.xlsx", index=False)
```
